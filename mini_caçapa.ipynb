```python
from pathlib import Path
import zipfile
import os
my_dir = Path("/content/drive/MyDrive/Caçapa validation/" )
out_dir = Path("/content/drive/MyDrive/Caçapa validation/training2" )

for i in range(1,522):
  my_zip = "/content/drive/MyDrive/Caçapa validation/train%.3d.zip" % (i)
  print("Descompactando %s" % (my_zip))
  zip_file = zipfile.ZipFile(my_zip, 'r')
  for files in zip_file.namelist():
    print(files)
    data = zip_file.read(files)
    myfile_path = out_dir/Path(files).name
    #print(myfile_path)
    if Path(myfile_path).is_file():
      print ("%s já existe"%(myfile_path))
      continue
    myfile_path.write_bytes(data)
  zip_file.close()
```
